# Import libraries

```python
import pandas as pd
import numpy as np
```

# Load data

```python
# Load initial data extract
# PSEI data
psei = pd.read_csv('PSEI_daily.csv', parse_dates=['Date'])

# Retain only the 'Date' and 'Close' column
psei = psei[['Date','Close']]
psei.rename(columns={'Close': 'PSEI_Close'}, inplace=True)
display(psei.head())
# Sentiment scores data
sentiment = pd.read_csv('Sentiment_scores.csv', parse_dates=['date'])

# Retain only the 'date' and sentiment score column
sentiment = sentiment[['date','negative_score','neutral_score','positive_score']]
sentiment.rename(columns={'date': 'Date'}, inplace=True)
display(sentiment.head())
```

```python
# Generate only weekdays (business days)
dates = pd.date_range(start=psei['Date'].min(), end=psei['Date'].max())

# Put into a DataFrame
df = pd.DataFrame({'Date': dates})
display(df.head(10))
display(df.tail(10))
```

# Merge data

```python
# Merge sentiment scores with the weekday DataFrame
df = df.merge(sentiment, on='Date', how='left')
display(df.head(10))
display(df.tail(10))
```

```python
# Merge PSEI data
df = df.merge(psei, left_on='Date', right_on='Date', how='left')
display(df.head(10))
display(df.tail(10))
```

# Adjust dates

```python
# Replace non-trading date to next trading date
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# Replace missing PSEI dates with NaT
df['Adjusted_Date'] = df['Date'].where(df['PSEI_Close'].notna())

# Backward fill to move missing dates to next non-missing date
df['Adjusted_Date'] = df['Adjusted_Date'].bfill()

# Sanity check after date adjustment
display(df.head(60))
```

# Fill missing data

```python
# Using PSEI data, replace Saturday and Sunday NaN values with next Monday's values
df['PSEI_Close'] = df.groupby('Adjusted_Date')['PSEI_Close'].transform('first')

# Fill missing sentiment scores with 0
df[['negative_score','neutral_score','positive_score']] = df[['negative_score','neutral_score','positive_score']].fillna(0)

# Sanity check after filling missing data
display(df.iloc[380:420])
```

# Define new variables

```python
# List of aggregation functions and corresponding suffixes
agg_funcs = {
    'mean': 'mean',
    'std': 'sd',
    'max': 'max',
    'min': 'min'
}

# Columns to aggregate
score_cols = ['negative_score', 'neutral_score', 'positive_score']

# Loop over aggregation functions
for func_name, suffix in agg_funcs.items():
    # Compute daily aggregation
    daily_agg = df.groupby('Adjusted_Date', as_index=False)[score_cols].agg(func_name)
    
    # Rename columns with suffix
    daily_agg = daily_agg.rename(columns={col: f"{col}_{suffix}" for col in score_cols})
    
    # Merge back to original df
    df = df.merge(daily_agg, on='Adjusted_Date', how='left')

# Replace missing SD to 0
df[['negative_score_sd', 'neutral_score_sd', 'positive_score_sd']] = df[['negative_score_sd', 'neutral_score_sd', 'positive_score_sd']].fillna(0)

# Sanity check after merging
display(df.head())
```

# Clean data for modeling

```python
# Remove variables that are needed for modeling
df = df.drop(columns=df.columns[[0,1,2,3]])

# Rearrange columns
cols = ['Adjusted_Date'] + [col for col in df.columns if col != 'Adjusted_Date']
df = df[cols]

# Retain one row per day
df = df.drop_duplicates()

# Check duplicate dates
print(df['Adjusted_Date'].value_counts()[lambda x: x > 1])

# Arrange in chronological order
df = df.sort_values(by='Adjusted_Date')

# Reset index
df.reset_index(drop=True, inplace=True)

# Sanity check for final df
display(df.head())
```

# Save data

```python
# Save dataframe locally
df.to_csv("df_model.csv", index = False)
```
